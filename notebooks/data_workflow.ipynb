**"Step-3":**
Read the .wav file and output its shape and sampling rate.

```python
import librosa

audio, sr = librosa.load("Bluewhale.wav", sr=22050)

print(audio.shape)
print(sr)
```

**"Step-4":**
Listen to the audio file.

```python
from IPython.display import Audio

Audio(audio, rate=sr)
```

**"Step-5:"**
Visualising the waveform

```python
plt.figure(figsize=(10,4))
plt.plot(audio)
plt.show()

#librosa.display.waveshow(audio, sr=sr)
#using librosa
```

**"Step-6:"**
Plotting a spectogram, youll immediately be able to tell where the energy is conserved.

```python
D = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max)

plt.figure(figsize=(10,5))
librosa.display.specshow(D,
                         sr=sr,
                         x_axis='time',
                         y_axis='log')

plt.colorbar()
plt.show()
```

**"Step-7:"**
Extracting the MFCCs.

```python
mfcc = librosa.feature.mfcc(y=audio,
                            sr=sr,
                            n_mfcc=13)

print(mfcc.shape)
```

**"Step-8:"**
Visualising the MFCCs.

```python
plt.figure(figsize=(10,5))

librosa.display.specshow(mfcc,
                         x_axis='time')

plt.colorbar()
plt.show()
```
